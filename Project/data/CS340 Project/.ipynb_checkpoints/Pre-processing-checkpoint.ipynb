```python
from fairlearn.preprocessing import CorrelationRemover
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./Data/train.csv')
print('loaded %d records' % len(train))
# for column in train.columns:
#     print(column)

# Make sure all comment_text values are strings
train['comment_text'] = train['comment_text'].astype(str) 

# List all identities
identity_columns = [
    'muslim', 'black','severe_toxicity', 'obscene', 'identity_attack', 'insult', 'threat']

# Convert taget and identity columns to booleans
def convert_to_bool(df, col_name):
    df[col_name] = np.where(df[col_name] >= 0.5, True, False)
    
def convert_dataframe_to_bool(df):
    bool_df = df.copy()
    for col in ['target'] + identity_columns:
        convert_to_bool(bool_df, col)
    return bool_df
train = convert_dataframe_to_bool(train)
```

```python
print(train.head(3))
test = pd.read_csv('./Data/test.csv')
print(test.head(3))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 计算相关系数矩阵
correlation_matrix = train[identity_columns].corr()

# 绘制热图
plt.figure(figsize=(110, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Boolean Features')
plt.show()
```

```python
# 选择身份属性列
identity_data = train[identity_columns]
# 使用 CorrelationRemover 消除 muslim 和 black 与其他属性的相关性
cr = CorrelationRemover(sensitive_feature_ids=['muslim', 'black'])
cr.fit(identity_data)
identity_data_transformed = cr.transform(identity_data)




# 转换后的数据重新生成 DataFrame，首先需要知道哪些列被去除了
remaining_columns = [col for col in identity_columns if col not in ['muslim', 'black']]

# 转换后的数据重新生成 DataFrame
identity_data_transformed_df = pd.DataFrame(identity_data_transformed, columns=remaining_columns)

# 手动将被去除的敏感特征添加回来
identity_data_transformed_df['muslim'] = identity_data['muslim']
identity_data_transformed_df['black'] = identity_data['black']

# 确保 'muslim' 和 'black' 列出现在 DataFrame 的开头
desired_column_order = ['muslim', 'black'] + remaining_columns
identity_data_transformed_df = identity_data_transformed_df.reindex(columns=desired_column_order)


# 打印消除相关性后的相关性矩阵
transformed_correlation_matrix = identity_data_transformed_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(transformed_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Transformed Correlation Heatmap of Boolean Features')
plt.show()
```

```python
# 将最终的 DataFrame 保存为 CSV 文件
identity_data_transformed_df.to_csv('transformed_identity_data.csv', index=False)
```
